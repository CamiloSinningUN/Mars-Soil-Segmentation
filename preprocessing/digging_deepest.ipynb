```python
import numpy as np

dataset_cleaned = np.load('../data/mars_trainning_set_clean_no_aliens.npz', allow_pickle=True)
X_train = dataset_cleaned['images']
y_train = dataset_cleaned['masks']

dataset_path = '../data/mars_for_students.npz'

data = np.load(dataset_path)
X_test = data["test_set"]
```

```python
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt

# Perform PCA
pca = PCA(n_components=2)
y_train_pca = pca.fit_transform(y_train.reshape(len(y_train), -1))

# Plot the PCA result
plt.scatter(y_train_pca[:, 0], y_train_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of y_train')
plt.show()
```

```python
# Find indices where class 4 is present in y_train
class_4_indices = np.where(y_train == 4)[0]

# Plot the PCA result
plt.scatter(y_train_pca[:, 0], y_train_pca[:, 1], label='Other Classes')
plt.scatter(y_train_pca[class_4_indices, 0], y_train_pca[class_4_indices, 1], color='red', label='Class 4')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of y_train with Class 4 Highlighted')
plt.legend()
plt.show()
```

```python
# Perform PCA on the images
X_train_flat = X_train.reshape(len(X_train), -1)
pca_images = PCA(n_components=2)
X_train_pca = pca_images.fit_transform(X_train_flat)

# Plot the PCA result
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], label='Other Classes')
plt.scatter(X_train_pca[class_4_indices, 0], X_train_pca[class_4_indices, 1], color='red', label='Class 4')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of X_train with Class 4 Highlighted')
plt.legend()
plt.show()
```

```python
# Perform PCA on the test images
X_test_flat = X_test.reshape(len(X_test), -1)
pca_test_images = PCA(n_components=2)
X_test_pca = pca_test_images.fit_transform(X_test_flat)

# Plot the PCA result
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of X_test')
plt.show()
```
